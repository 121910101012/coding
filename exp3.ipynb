```python
#Find the permutation of list:['Rama','Krishna','Hari','You','And','Me']

from itertools import permutations
s=['Rama','Krishna','Hari','You','And','Me']
result=list(permutations(s))
for i in result:
  print(i)
m=len(result)
print("Number of possible ways:",m)
```

```python
#find the permutation of list:[12.09,"R",'K','3']

from itertools import permutations
s=[12.09,"R",'K','3']
result=list(permutations(s))
for i in result:
  print(i)
m=len(result)
print("The number of possible ways:",m)
```
